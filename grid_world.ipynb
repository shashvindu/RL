```python
import numpy as np

# Define the grid size
GRID_SIZE = 5

# Define actions
actions = ['up', 'down', 'left', 'right']

# Initialize agent position
agent_pos = [0, 0]  # top-left corner

# Define goal position
goal_pos = [5, 5]  # bottom-right corner

def print_grid(agent_pos, goal_pos):
    for i in range(GRID_SIZE):
        row = ''
        for j in range(GRID_SIZE):
            if [i,j] == agent_pos:
                row += 'A '  # Agent
            elif [i,j] == goal_pos:
                row += 'G '  # Goal
            else:
                row += '. '  # Empty cell
        print(row)
    print("\n")

def move_agent(agent_pos, action):
    if action == 'up' and agent_pos[0] > 0:
        agent_pos[0] -= 1
    elif action == 'down' and agent_pos[0] < GRID_SIZE - 1:
        agent_pos[0] += 1
    elif action == 'left' and agent_pos[1] > 0:
        agent_pos[1] -= 1
    elif action == 'right' and agent_pos[1] < GRID_SIZE - 1:
        agent_pos[1] += 1
    return agent_pos

# Game loop
while agent_pos != goal_pos:
    print_grid(agent_pos, goal_pos)
    print("Available actions:", actions)
    action = input("Enter action: ")
    if action in actions:
        agent_pos = move_agent(agent_pos, action)
    else:
        print("Invalid action. Try again!")

print_grid(agent_pos, goal_pos)
print("Congratulations! You reached the goal!")
```
